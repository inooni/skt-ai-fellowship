```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
 
 
 
dataset = pd.read_csv('/content/drive/My Drive/NASA_Bearing/2nd_test/2nd_test/2004.02.18.18.52.39',sep='\t',header = None)
 
dataset.columns = ['Bearing 1', 'Bearing 2', 'Bearing 3', 'Bearing 4']
 
signal = list(dataset['Bearing 1'].values)
 
sum = signal[:]
for i in range(1,len(sum)):
    sum[i] = sum[i] + sum[i-1]
 
sumsum = sum[:]
for i in range(1,len(sumsum)):
    sumsum[i] = sumsum[i] + sumsum[i-1]
 
 
fs = len(signal)
 
t = np.arange(0,1,1/fs)
 
plt.subplot(3,1,1)
plt.plot(t,signal)
plt.grid()
 
plt.subplot(3,1,2)
plt.plot(t,[x / 20480 for x in sum])
plt.grid()
 
plt.subplot(3,1,3)
plt.plot(t,[x / (20480*20480) for x in sumsum])
plt.grid()
 
plt.show()
```
